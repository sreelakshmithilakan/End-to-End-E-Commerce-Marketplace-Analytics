```python
#final customer segment column.
def segment_customer(row):

    if row['R_score'] >= 4 and row['M_score'] >= 4:
        return 'Champion'

    elif row['R_score'] >= 4 and row['M_score'] <= 3:
        return 'Potential Loyalist'

    elif row['R_score'] <= 2 and row['M_score'] >= 4:
        return 'At Risk'

    elif row['R_score'] <= 2 and row['M_score'] <= 2:
        return 'Lost Customer'

    else:
        return 'Regular Customer'
```

```python
rfm['customer_segment'] = rfm.apply(
    segment_customer,
    axis=1
)
```

# Visualization

```python
import plotly.express as px
```

```python
#1. Customer Value Segmentation
segment_counts = (
    rfm['customer_segment']
    .value_counts()
    .reset_index()
)

segment_counts.columns = ['Segment', 'Customers']

fig = px.bar(
    segment_counts,
    x='Customers',
    y='Segment',
    orientation='h',
    color='Segment',
    title='Customer Value Segmentation',
    color_discrete_sequence=px.colors.qualitative.Bold
)

fig.update_traces(
    marker_line_color='black',
    marker_line_width=1
)

fig.update_layout(
    title_x=0.5,
    plot_bgcolor='white',
    paper_bgcolor='white',
    xaxis_title='Number of Customers',
    yaxis_title='',
    showlegend=False
)

fig.show()
```

```python
# 2. RFM Heatmap


heatmap = (
    rfm.groupby(['R_score', 'M_score'])
       .size()
       .reset_index(name='Customers')
)

fig = px.density_heatmap(
    heatmap,
    x='M_score',
    y='R_score',
    z='Customers',
    text_auto=True,
    color_continuous_scale='RdYlGn',
    title='RFM Customer Distribution Heatmap'
)

fig.update_layout(
    title_x=0.5,
    plot_bgcolor='white',
    paper_bgcolor='white',
    xaxis_title='Monetary Score',
    yaxis_title='Recency Score'
)

fig.show()
```

```python
# 3: Average Spend by Customer Segment
avg_spend = (
    rfm.groupby('customer_segment')['monetary']
       .mean()
       .sort_values(ascending=False)
       .reset_index()
)

fig = px.bar(
    avg_spend,
    x='customer_segment',
    y='monetary',
    color='customer_segment',
    title='Average Spend by Customer Segment',
    color_discrete_sequence=px.colors.qualitative.Set2
)

fig.update_traces(
    marker_line_color='black',
    marker_line_width=1
)

fig.update_layout(
    title_x=0.5,
    plot_bgcolor='white',
    paper_bgcolor='white',
    xaxis_title='Customer Segment',
    yaxis_title='Average Spend',
    showlegend=False
)

fig.show()
```

```python
#4: Frequency Segment Distribution
retention_counts = (
    rfm['frequency_segment']
    .value_counts()
    .reset_index()
)

retention_counts.columns = ['Segment', 'Customers']

fig = px.bar(
    retention_counts,
    x='Segment',
    y='Customers',
    color='Segment',
    title='Frequency Segment Distribution',
    color_discrete_map={
        'One-Time Buyer': '#E63946',
        'Returning Customer': '#457B9D',
        'Loyal Customer': '#2A9D8F'
    }
)

fig.update_traces(
    marker_line_color='black',
    marker_line_width=1
)

fig.update_layout(
    title_x=0.5,
    plot_bgcolor='white',
    paper_bgcolor='white',
    yaxis_title='Number of Customers',
    showlegend=False
)

fig.show()
```

```python
#5. Customer Segment vs Retention Segment
cross_tab = pd.crosstab(
    rfm['customer_segment'],
    rfm['frequency_segment']
)

fig = px.imshow(
    cross_tab,
    text_auto=True,
    color_continuous_scale='Blues',
    aspect='auto',
    title='Customer Value Segment vs Retention Segment'
)

fig.update_layout(
    title_x=0.5,
    plot_bgcolor='white',
    paper_bgcolor='white'
)

fig.show()
```
